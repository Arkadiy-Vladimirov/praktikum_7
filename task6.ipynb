$$ y_{i_1,i_2,\ldots,i_{10}} = \left( \dfrac{t_{i_1} - t_{i_1-1}}{2},

i_j = \overline{1,n},\: \forall j = \overline{1,10} \}$
and contains $n^{10}$ nodes. Finally
$$\widehat I_n = \frac{1}{n^{10}} \sum_{m \in M} f(m).$$
